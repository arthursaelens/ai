# Encoding

## Pre-processing and selecting the metric

```python
import os
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.layers import Activation, Dense, Input
from tensorflow.keras.layers import Conv2D, Flatten
from tensorflow.keras.layers import Reshape, Conv2DTranspose
from tensorflow.keras.layers import Cropping2D
from tensorflow.keras.models import Model
from tensorflow.keras import layers, models
from tensorflow.keras import backend as K
```

```python
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
```

```python
np.random.seed(1337)
tf.random.set_seed(1337)
```

```python
# Load Fashion MNIST dataset
fashion_mnist = tf.keras.datasets.fashion_mnist
(all_train_images, _), (test_images, _) = fashion_mnist.load_data()

image_size = all_train_images.shape[1]

# Check dataset shape
print("Training data shape:", all_train_images.shape)
print("Testing data shape:", test_images.shape)
```

```python
all_train_images = np.reshape(all_train_images, (-1, image_size, image_size, 1))
all_train_images = all_train_images.astype('float32') / 255

test_images = np.reshape(test_images, (-1, image_size, image_size, 1))
test_images = test_images.astype('float32') / 255
```

```python
# Split the original training data into new training and validation sets
train_images, train_images_validation, train_labels, train_labels_validation = train_test_split(all_train_images, all_train_labels, test_size=0.2, random_state=42)

print("Training set size:", train_images.shape)
print("Validation set size:", train_images_validation.shape)
print("Test set size:", test_images.shape)
```

```python
# Making noisy images so we can test our model

noise = np.random.normal(loc=0.5, scale=0.5, size=all_train_images.shape)
all_train_images_noisy = all_train_images + noise

noise = np.random.normal(loc=0.5, scale=0.5, size=train_images.shape)
train_images_noisy = train_images + noise

noise = np.random.normal(loc=0.5, scale=0.5, size=train_images_validation.shape)
train_images_validation_noisy = train_images_validation + noise

noise = np.random.normal(loc=0.5, scale=0.5, size=test_images.shape)
test_images_noisy = test_images + noise

# We will cut off the values that fell out of the range [0,1] after adding the noise
train_images_noisy = np.clip(train_images_noisy, 0., 1.)
test_images_noisy = np.clip(test_images_noisy, 0., 1.)

print("Training data shape:", all_train_images_noisy.shape)
print("Training data shape:", train_images_noisy.shape)
print("Training data shape:", train_images_validation_noisy.shape)
print("Testing data shape:", test_images_noisy.shape)
```

```python
loss = 'mse'
```

# Train our initial model

```python
input_shape = (image_size, image_size, 1)
kernel_size = 3
layer_filters = [32, 64, 128]
```

### Encoder

```python
# We have the input layer
inputs = Input(shape=input_shape, name='encoder_input')
x = inputs

# Afterwards we have two convolutional layers
for filters in layer_filters:
    x = Conv2D(filters=filters,
               kernel_size=kernel_size,
               strides=2,
               activation='relu',
               padding='same')(x)

# Final convolutional layer to reach the latent space shape of (4, 4, 32)
x = Conv2D(filters=32, kernel_size=kernel_size, strides=1, activation='relu', padding='same')(x)

# Now `x` has shape (None, 4, 4, 32), which is our desired latent space
latent_outputs = x

encoder = Model(inputs=inputs, outputs=latent_outputs, name='encoder')
encoder.summary()
```

### Decoder

```python
# First we have the input layer
latent_inputs = Input(shape=(4, 4, 32), name='decoder_input')
x = latent_inputs

# then we add transposed convolutional layers but in a reversed order compared to the encoder
for filters in layer_filters[::-1]:
    x = Conv2DTranspose(filters=filters,
                    kernel_size=kernel_size,
                    strides=2,
                    activation='relu',
                    padding='same')(x)

x = Conv2DTranspose(1, kernel_size=kernel_size, strides=1, activation='sigmoid', padding='same', name='decoder_output')(x)

outputs = x = Cropping2D(((2, 2), (2, 2)))(x) 

decoder = Model(inputs=latent_inputs, outputs=outputs, name='decoder')
decoder.summary(line_length=110)
```

### Autoencoder

```python
autoencoder = Model(inputs, decoder(encoder(inputs)), name='autoencoder')
autoencoder.summary()
```

```python
autoencoder.compile(loss=loss, optimizer='adam')
```

```python
epochs = 10
batch_size = 128

autoencoder.fit(train_images_noisy,
                train_images,
                validation_data=(train_images_validation_noisy, train_images_validation),
                epochs=epochs,
                batch_size=batch_size)
```

First we'll try to increase the filter

```python
loss = 'mse'
```

```python
# Updated hyperparameters
layer_filters = [64, 128, 256]

# Rebuild the encoder with the updated filters
inputs = Input(shape=input_shape, name='encoder_input')
x = inputs

for filters in layer_filters:
    x = Conv2D(filters=filters,
               kernel_size=kernel_size,
               strides=2,
               activation='relu',
               padding='same')(x)

x = Conv2D(filters=32, kernel_size=kernel_size, strides=1, activation='relu', padding='same')(x)
latent_outputs = x
encoder_opt = Model(inputs=inputs, outputs=latent_outputs, name='encoder_opt')

# Rebuild the decoder
latent_inputs = Input(shape=(4, 4, 32), name='decoder_input')
x = latent_inputs
for filters in layer_filters[::-1]:
    x = Conv2DTranspose(filters=filters,
                    kernel_size=kernel_size,
                    strides=2,
                    activation='relu',
                    padding='same')(x)

x = Conv2DTranspose(1, kernel_size=kernel_size, strides=1, activation='sigmoid', padding='same', name='decoder_output')(x)
outputs = Cropping2D(((2, 2), (2, 2)))(x)
decoder_opt = Model(inputs=latent_inputs, outputs=outputs, name='decoder_opt')
```

```python
# Build the autoencoder
autoencoder_opt_mse = Model(inputs, decoder_opt(encoder_opt(inputs)), name='autoencoder_opt')

autoencoder_opt_mse.summary()

# Compile the model
autoencoder_opt_mse.compile(optimizer='adam', loss=loss)

# Train the model
history = autoencoder_opt_mse.fit(train_images_noisy, train_images, epochs=10, validation_data=(train_images_validation_noisy, train_images_validation))
```

## Decrease the latent space

```python
# Reduced feature dim to 28
```

```python
layer_filters = [128, 256, 512]

# Rebuild the encoder with reduced feature dimension (32 to 28 filters)
inputs = Input(shape=input_shape, name='encoder_input')
x = inputs

for filters in layer_filters:
    x = Conv2D(filters=filters,
               kernel_size=kernel_size,
               strides=2,
               activation='relu',
               padding='same')(x)

# Reduce feature dimension in the final layer from 32 to 28 (decrease by 4)
x = Conv2D(filters=28, kernel_size=kernel_size, strides=1, activation='relu', padding='same')(x)
latent_outputs = x  # Now output shape will be (4, 4, 28)
encoder_opt = Model(inputs=inputs, outputs=latent_outputs, name='encoder_opt')

# Rebuild the decoder to match the new encoder output shape
latent_inputs = Input(shape=(4, 4, 28), name='decoder_input')
x = latent_inputs

for filters in layer_filters[::-1]:
    x = Conv2DTranspose(filters=filters,
                        kernel_size=kernel_size,
                        strides=2,
                        activation='relu',
                        padding='same')(x)

# Final layer to match the original input channels (1 for grayscale, or adjust as needed)
x = Conv2DTranspose(1, kernel_size=kernel_size, strides=1, activation='sigmoid', padding='same', name='decoder_output')(x)
outputs = Cropping2D(((2, 2), (2, 2)))(x)
decoder_opt = Model(inputs=latent_inputs, outputs=outputs, name='decoder_opt')

autoencoder_decr_lat = Model(inputs, decoder_opt(encoder_opt(inputs)), name='autoencoder_opt')

autoencoder_decr_lat.compile(optimizer='adam', loss=loss)

autoencoder_decr_lat.summary()

# Train the model
history = autoencoder_decr_lat.fit(train_images_noisy, train_images, epochs=10, validation_data=(test_images_noisy, test_images))  # dont forget to change to train_images_validation_noisy
```
